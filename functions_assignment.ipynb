```python
#ans1
def sum_of_even(l):
  sum=0
  for i in l:
    if i%2==0:
      sum+=i
  return sum

user_input=input("enter the list of numbers: ")
l1=list(map(int,user_input.split()))
print(sum_of_even(l1))
```

```python
#ans 2
def reverse_string(rv):
  a=len(rv)-1
  l=[]
  while a>=0:
     l.append(rv[a])
     a-=1
  return str(l)


s=input("enter an string : ")
reverse_string(s)
```

```python
#ans3
def sqaure(l):
  l1=[]
  for i in l:
    l1.append(i**2)
  return l1

st=input("enter the list of numbers: ")
l=list(map(int,st.split()))
sqaure(l)
```

```python
#ans4
def check_prime(n):
  if n<=1:
    return False
  for i in range(2,n):
    if n%i==0 :
      return False
  return True


for i in range(1,201):
  check_prime(i)
```

```python
#ans 5
def fib(n):
  a=0
  b=1
  for i in range(n+1):
    yield a
    a,b=b,a+b

c=list(fib(10))
c
```

```python
#ans6
def exp(n):
  for i in range(1,n+1):
    yield 2**i

a=list(exp(5))
a
```

```python
#ans7
def read_file_line_by_line(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()
```

```python
#ans 10
def f(input_string):
  vowels="aeiouAEIOU"
  a=filter(lambda x:x not in vowels,input_string)
  return "".join(a)

s=input("enter the string: ")
f(s)
```
